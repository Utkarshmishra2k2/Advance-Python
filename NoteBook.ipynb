 Write a Python program to evaluate a polynomial for a given value of x using Horner’s Method. Compare its efficiency with the naive approach.

image.png

```python
def horner_eval(poly, x):
    result = poly[0]
    for i in range(1, len(poly)):
        result = result * x + poly[i]
    return result
```

```python
def naive_eval(poly, x):
    result = 0
    n = len(poly) - 1
    for i, coeff in enumerate(poly):
        result += coeff * (x ** (n - i))
    return result
```

```python
print(f"Polynomial evaluation for x=3 :")
print(f"Horner's Method: {horner_eval([2, -6, 2, -1], 3)}")
print(f"Naive Method:   {naive_eval([2, -6, 2, -1], 3)}")
```

Create a program that identifies all Pythagorean triplets within a given range, but with the constraint that the triplet must also satisfy an additional condition: the sum of the triplet must be a prime number.

```python
import math
data = int(input("Enter the number : "))
def is_prime(x):
    if x < 2:
        return False
    for i in range(2, int(math.sqrt(x)) + 1):
        if x % i == 0:
            return False
    return True

triplets = []

for m in range(1, data+1):
    for n in range(1, m):
        a = m*m - n*n
        b = 2*m*n
        c = m*m + n*n
        s = a + b + c

        if is_prime(s):
            triplets.append((a, b, c))

print(triplets)
```

```python
principal=500000
annual_rate=10
tenure_months=60
emi=11000
```

```python
monthly_rate = annual_rate / (12 * 100)
balance = principal
schedule = []
early_payoff_month = None
for month in range(1, tenure_months + 1):
    if balance <= 0:
        break

    interest = balance * monthly_rate
    principal_paid = emi - interest
    if principal_paid <= 0:
        print("EMI is too low. Loan will never be repaid.")
        print(schedule, None)

    balance -= principal_paid
    if balance < 0:
        principal_paid += balance
        emi_actual = interest + principal_paid
        balance = 0
    else:
        emi_actual = emi
    schedule.append({
        "month": month,
        "emi_paid": round(emi_actual, 2),
        "interest": round(interest, 2),
        "principal_paid": round(principal_paid, 2),
        "remaining_balance": round(balance, 2)
    })

    if balance == 0:
        early_payoff_month = month
        break
print(schedule, early_payoff_month)
```

[Link](https://www.sanfoundry.com/python-program-find-area-triangle-given-three-sides/)

```python
import math

def area_base_height():
    base = float(input("Enter the base of the triangle: "))
    height = float(input("Enter the height of the triangle: "))
    area = 0.5 * base * height
    print(f"Area of the triangle = {area}")

def area_heron():
    a = float(input("Enter side a: "))
    b = float(input("Enter side b: "))
    c = float(input("Enter side c: "))
    if a + b > c and a + c > b and b + c > a:
        s = (a + b + c) / 2
        area = math.sqrt(s * (s - a) * (s - b) * (s - c))
        print(f"Area of the triangle = {area}")
    else:
        print("Invalid triangle sides!")

def area_coordinates():
    x1 = float(input("Enter x1: "))
    y1 = float(input("Enter y1: "))
    x2 = float(input("Enter x2: "))
    y2 = float(input("Enter y2: "))
    x3 = float(input("Enter x3: "))
    y3 = float(input("Enter y3: "))
    area = abs((x1*(y2 - y3) + x2*(y3 - y1) + x3*(y1 - y2)) / 2)
    print(f"Area of the triangle = {area}")

def main():
    print("Choose the method to calculate the area of a triangle:")
    print("1. Using Base and Height")
    print("2. Using Heron's Formula (3 sides)")
    print("3. Using Coordinates")

    choice = input("Enter your choice (1/2/3): ")

    if choice == '1':
        area_base_height()
    elif choice == '2':
        area_heron()
    elif choice == '3':
        area_coordinates()
    else:
        print("Invalid choice!")

main()
```

**1948. Delete Duplicate Folders in System**

[Link](https://leetcode.com/problems/delete-duplicate-folders-in-system/description/?envType=daily-question&envId=2025-07-20)

```python
class Trie:
    # current node structure's serialized representation
    serial: str = ""
    # current node's child nodes
    children: dict

    def __init__(self):
        self.children = dict()


class Solution:
    def deleteDuplicateFolder(self, paths: List[List[str]]) -> List[List[str]]:
        # root node
        root = Trie()

        for path in paths:
            cur = root
            for node in path:
                if node not in cur.children:
                    cur.children[node] = Trie()
                cur = cur.children[node]

        # hash table records the occurrence times of each serialized representation
        freq = Counter()

        # post-order traversal based on depth-first search, calculate the serialized representation of each node structure
        def construct(node: Trie) -> None:
            # if it is a leaf node, then the serialization is represented as an empty string, and no operation is required.
            if not node.children:
                return

            v = list()
            # if it is not a leaf node, the serialization representation of the child node structure needs to be calculated first.
            for folder, child in node.children.items():
                construct(child)
                v.append(folder + "(" + child.serial + ")")

            # to prevent issues with order, sorting is needed
            v.sort()
            node.serial = "".join(v)
            # add to hash table
            freq[node.serial] += 1

        construct(root)

        ans = list()
        # record the path from the root node to the current node.
        path = list()

        def operate(node: Trie) -> None:
            # if the serialization appears more than once in the hash table, it needs to be deleted.
            if freq[node.serial] > 1:
                return
            # otherwise add the path to the answer
            if path:
                ans.append(path[:])

            for folder, child in node.children.items():
                path.append(folder)
                operate(child)
                path.pop()

        operate(root)
        return ans
```

**K-th Smallest in Lexicographical Order**

[Link](https://leetcode.com/problems/k-th-smallest-in-lexicographical-order/description/?envType=daily-question&envId=2025-06-09)

```python
class Solution(object):
    def findKthNumber(self, n, k):
        curr = 1
        k -= 1

        while k > 0:
            step = self._count_steps(n, curr, curr + 1)
            if step <= k:
                curr += 1
                k -= step
            else:
                curr *= 10
                k -= 1

        return curr

    def _count_steps(self, n, prefix1, prefix2):
        steps = 0
        while prefix1 <= n:
            steps += min(n + 1, prefix2) - prefix1
            prefix1 *= 10
            prefix2 *= 10
        return steps
```

**3335. Total Characters in String After Transformations I**

[Link](https://leetcode.com/problems/total-characters-in-string-after-transformations-i/description/?envType=daily-question&envId=2025-05-13)

```python
class Solution:
    def lengthAfterTransformations(self, s: str, t: int) -> int:
        mod = 10**9 + 7
        cnt = [0] * 26
        for ch in s:
            cnt[ord(ch) - ord("a")] += 1
        for round in range(t):
            nxt = [0] * 26
            nxt[0] = cnt[25]
            nxt[1] = (cnt[25] + cnt[0]) % mod
            for i in range(2, 26):
                nxt[i] = cnt[i - 1]
            cnt = nxt
        ans = sum(cnt) % mod
        return ans
```

**2594. Minimum Time to Repair Cars**

[Link](https://leetcode.com/problems/minimum-time-to-repair-cars/description/?envType=daily-question&envId=2025-03-16)

```python
import math
from typing import List

class Solution:
    def repairCars(self, ranks: List[int], cars: int) -> int:
        def canRepairInTime(time):
            total_cars_repaired = 0
            for r in ranks:
                total_cars_repaired += int(math.sqrt(time / r))
                if total_cars_repaired >= cars:
                    return True
            return total_cars_repaired >= cars
        low, high = 1, min(ranks) * cars * cars
        while low < high:
            mid = (low + high) // 2
            if canRepairInTime(mid):
                high = mid
            else:
                low = mid + 1

        return low
```

**875. Koko Eating Bananas**

[Link](https://leetcode.com/problems/koko-eating-bananas/description/)

```python
class Solution:
    def minEatingSpeed(self, piles: List[int], h: int) -> int:
        def canFinish(k):
            total_hours = 0
            for pile in piles:
                total_hours += math.ceil(pile / k)
                if total_hours > h:
                    return False
            return total_hours <= h
        low, high = 1, max(piles)

        while low < high:
            mid = (low + high) // 2
            if canFinish(mid):
                high = mid  # Try to find a smaller k
            else:
                low = mid + 1  # Increase k

        return low
```

**Minimum Operations to Make a Uni-Value Grid**

[Link](https://leetcode.com/problems/minimum-operations-to-make-a-uni-value-grid/description/?envType=daily-question&envId=2025-03-26)

```python
from typing import List
class Solution:
    def minOperations(self, grid: List[List[int]], x: int) -> int:
        flat_grid = [num for row in grid for num in row]
        remainder = flat_grid[0] % x
        for num in flat_grid:
            if num % x != remainder:
                return -1
        normalized_grid = [(num - remainder) // x for num in flat_grid]
        normalized_grid.sort()
        median = normalized_grid[len(normalized_grid) // 2]
        operations = sum(abs(num - median) for num in normalized_grid)
        return operations
```

**21. Merge Two Sorted Lists**

[Link](https://leetcode.com/problems/merge-two-sorted-lists/)

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode()
        current = dummy
        while list1 and list2:
            if list1.val <= list2.val:
                current.next = list1
                list1 = list1.next
            else:
                current.next = list2
                list2 = list2.next
            current = current.next
        current.next = list1 if list1 else list2

        return dummy.next
```

**763. Partition Labels**

[Link](https://leetcode.com/problems/partition-labels/description/?envType=daily-question&envId=2025-03-30)

```python
class Solution:
    def partitionLabels(self, s: str) -> List[int]:
        last_occurrence = {char: idx for idx, char in enumerate(s)}
        result = []
        start, end = 0, 0
        for idx, char in enumerate(s):
            end = max(end, last_occurrence[char])
            if idx == end:
                result.append(idx - start + 1)
                start = idx + 1

        return result
```

**368. Largest Divisible Subset**

[Link](https://leetcode.com/problems/largest-divisible-subset/description/?envType=daily-question&envId=2025-04-06)

```python
class Solution:
    def largestDivisibleSubset(self, nums: List[int]) -> List[int]:
        nums.sort()

        n = len(nums)
        dp = [1] * n
        prev = [-1] * n
        for i in range(1, n):
            for j in range(i):
                if nums[i] % nums[j] == 0 and dp[i] < dp[j] + 1:
                    dp[i] = dp[j] + 1
                    prev[i] = j
        max_size = max(dp)
        index = dp.index(max_size)
        result = []
        while index != -1:
            result.append(nums[index])
            index = prev[index]
        return result[::-1]
```

**416. Partition Equal Subset Sum**

[Link](https://leetcode.com/problems/partition-equal-subset-sum/description/?envType=daily-question&envId=2025-04-07)

```python
class Solution:
    def canPartition(self, nums: List[int]) -> bool:
        total_sum = sum(nums)
        if total_sum % 2 != 0:
            return False

        target = total_sum // 2
        dp = [False] * (target + 1)
        dp[0] = True
        for num in nums:
            for i in range(target, num - 1, -1):
                dp[i] = dp[i] or dp[i - num]
        return dp[target]
```

**657. Minimum number of stops**


You are provided with two sorted arrays, arrival and departure, containing the arrival and departure timings, respectively, of the bus at different destinations. Your task is to find the minimum number of stops required for the bus to reach the destination comfortably and ontime.

[Link](https://machinehack.com/practices/minimum-number-of-stops-be8e0/description)

```python
class Solution:
    def MinStopsReq(self, arrival, departure):
        if not arrival:
            return 0

        events = []
        for a in arrival:
            events.append((a, 'arrival'))
        for d in departure:
            events.append((d, 'departure'))
        events.sort(key=lambda x: (x[0], x[1] == 'arrival'))

        stops = 0
        current_passengers = 0
        for event in events:
            if event[1] == 'arrival':
                current_passengers += 1
                stops = max(stops, current_passengers)
            else:
                current_passengers -= 1

        return stops

solution = Solution()
print(solution.MinStopsReq(arrival, departure))
```

**643. Frequent digit**

You are provided with an array of numerical strings, arr. Return the most frequent digit(i.e.["202","20"] → 2). If there are two results for the most frequent digit, consider the first occurring number (i.e..["502", "205"] → 5 ).[Link](https://machinehack.com/practices/frequent-digit-45764/description)

```python
class Solution:
    def MostFreqDigit(self, arr):
        from collections import defaultdict

        digit_count = defaultdict(int)
        first_occurrence = {}

        for idx, num in enumerate(arr):
            for digit in num:
                if digit not in first_occurrence:
                    first_occurrence[digit] = idx
                digit_count[digit] += 1

        max_freq = -1
        result_digit = None

        for digit in digit_count:
            if digit_count[digit] > max_freq or (digit_count[digit] == max_freq and first_occurrence[digit] < first_occurrence[result_digit]):
                max_freq = digit_count[digit]
                result_digit = digit

        return int(result_digit)


# Driver code
solution = Solution()
print(solution.MostFreqDigit(arr))
```

**629. Remove even-length words**

After that, return True if the length of the new string is prime; otherwise, return False.[Link](https://machinehack.com/practices/remove-even-length-words-f26d9/description)

```python
class Solution:
    def is_prime(self, num):
        if num <= 1:
            return False
        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                return False
        return True

    def evenLengthWords(self, str1: str):
        words = str1.split()
        filtered_words = [word for word in words if len(word) % 2 != 0]
        new_string = ''.join(filtered_words)

        return self.is_prime(len(new_string))

obj = Solution()
print(obj.evenLengthWords(str1))
```

**174. Outliers in a dataset**

Consider data as data1 comprising n number of integers. Detect outliers in this given data using the IQR method.[Link](https://machinehack.com/practices/detecting-outliers-in-a-dataset-5bb3d/description)

```python
import numpy as np
def outlierDetection(data1):
# Write your code here
  data_sorted = sorted(data1)
  Q1 = np.percentile(data_sorted, 25)
  Q3 = np.percentile(data_sorted, 75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  outliers = [x for x in data1 if x < lower_bound or x > upper_bound]
  return outliers
# Driver Code
print(outlierDetection(data1))
```

Write a query identifying the type of each record in the TRIANGLES table using its three side lengths. Output one of the following statements for each record in the table:

Equilateral: It's a triangle with  sides of equal length.
Isosceles: It's a triangle with  sides of equal length.
Scalene: It's a triangle with  sides of differing lengths.
Not A Triangle: The given values of A, B, and C don't form a triangle.

image.png

[Link](https://www.hackerrank.com/challenges/what-type-of-triangle/problem?isFullScreen=true)

```python
SELECT
    CASE
    WHEN A+B<=C OR C+B<=A OR A+C<=B THEN "Not A Triangle"
    WHEN A = B AND B = C THEN "Equilateral"
    WHEN A = B OR A = C OR B = C THEN "Isosceles"
     ELSE "Scalene"
    END AS Triangle_Type
FROM
    TRIANGLES;
```
